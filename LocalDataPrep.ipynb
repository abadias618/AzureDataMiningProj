```python
import pandas as pd
import numpy as np
```

```python
#calculate total number of rows for function durations files
row_count = 0
for i in range(1,15):
    if i < 10:
        number = "0" + str(i)
    else:
        number = str(i)
    file = pd.read_csv("../azurefunctions-dataset2019/function_durations_percentiles.anon.d"+number+".csv")
    length = len(file)
    print("function_durations_percentiles.anon.d"+number+".csv --> "+str(length)+" rows")
    del file
    row_count += length
print("Total Rows",row_count)
```

```python
#calculate total number of rows for App memory percentiles files
row_count = 0
for i in range(1,13):
    if i < 10:
        number = "0" + str(i)
    else:
        number = str(i)
    file = pd.read_csv("../azurefunctions-dataset2019/app_memory_percentiles.anon.d"+number+".csv")
    length = len(file)
    print("invocations_per_function_md.anon.d"+number+".csv --> "+str(length)+" rows")
    del file
    row_count += length
print("Total Rows",row_count)
```

```python
# join all files into 2 dfs
gen_dura = pd.DataFrame()
gen_invo = pd.DataFrame()
invos_per_min_cols = [str(x) for x in range(1,1441)]

for i in range(1,15):
    if i < 10:
        number = "0" + str(i)
    else:
        number = str(i)
        
    dura = pd.read_csv("../azurefunctions-dataset2019/function_durations_percentiles.anon.d"+number+".csv")
    invo = pd.read_csv("../azurefunctions-dataset2019/invocations_per_function_md.anon.d"+number+".csv")
    
    invo = invo.drop(columns=invos_per_min_cols)
    gen_dura = pd.concat([gen_dura, dura])
    gen_invo = pd.concat([gen_invo, invo])
    
print(gen_dura.shape)
print(gen_invo.shape)
```

```python
#reset indexes
gen_dura = gen_dura.reset_index(drop=True)
gen_invo = gen_invo.reset_index(drop=True)
print(gen_dura.shape)
print(gen_invo.shape)
```

```python
#join and erase duplicates
joined = pd.merge(gen_dura, gen_invo, on=['HashOwner','HashFunction','HashApp'], how='inner')
joined = joined.drop_duplicates()
```

```python
#turn trigger column to numerical
joined = joined.reset_index(drop=True)
joined.replace(["http","timer","event","queue","storage","orchestration","others"],
                        [1,2,3,4,5,6,7], inplace=True)
```
